```python
import numpy as np
import scipy.io
import scipy.misc
import matplotlib.pyplot as plt
from scipy.special import expit
import matplotlib.cm as cm #Used to display images in a specific colormap
import random
```

```python
mat = scipy.io.loadmat("ex3data1.mat")
X = mat["X"]
y = mat["y"].ravel() # IMPORTANT! For some reason the code will not work if .ravel() is not used. I've yet to figure out exactly why this happens.
```

```python
# set up parameters
input_layer_size  = 400
num_labels = 10      
```

```python
def plotImage(image):
    width, height = 20,20
    return image.reshape(width,height).T
```

```python
def displayData(indices_to_display = None):
    """
    Function that picks 100 random rows from X, creates a 20x20 image from each,
    then stitches them together into a 10x10 grid of images, and shows it.
    """
    width, height = 20, 20
    nrows, ncols = 10, 10 # we want 100 digits to be displayed
    if not indices_to_display:
        indices_to_display = random.sample(range(X.shape[0]), nrows*ncols) # take random sample of 100 digits
        
    big_picture = np.zeros((height*nrows,width*ncols))
    
    irow, icol = 0, 0
    for idx in indices_to_display:
        if icol == ncols:
            irow += 1
            icol  = 0
        iimg = plotImage(X[idx])
        big_picture[irow*height:irow*height+iimg.shape[0],icol*width:icol*width+iimg.shape[1]] = iimg
        icol += 1
    fig = plt.figure(figsize=(6,6))
    img = scipy.misc.toimage( big_picture )
    plt.imshow(img,cmap = cm.Greys_r)
```

```python
displayData()
```

```python
# hypothesis function
def h(z):
    return expit(z)
```

```python
# set up test case
theta_t = np.matrix("-2; -1; 1; 2")
X_t = np.insert(np.matrix(np.arange(1,16)/10).reshape((5,3), order="F"),0,1,axis=1)
y_t = np.matrix("1;0;1;0;1")
lambda_t = 3
```

```python
X_t
```

```python
# Cost function J and gradient Regularized
def lr_cost_function(theta, X, y, lmda):
    m,n = X.shape
    J = 0
    grad = np.zeros(theta.shape)
    
    mask = np.eye(len(theta))
    mask[0,0] = 0
    
    reg = (lmda/(2*m))*np.sum(np.power(theta[1:len(theta)],2))
    J = (1.0/m)*np.sum(-y.T.dot(np.log(h(X.dot(theta))))-(1-y).T.dot(np.log(1-h(X.dot(theta))))) + reg
    
    X_dot_theta = X.dot(theta)
    grad = (1/m)*(h(X.dot(theta))-y).T.dot(X).T + (lmda/m)*(mask.dot(theta))
    
    return J, grad
```

```python
J, grad = lr_cost_function(theta_t,X_t,y_t,lambda_t)
```

```python
print("Cost function: ", J, "\n","Gradient:\n", grad)
```

```python
import scipy.optimize as spo
```

```python
# One-Vs-All classification --> train a classification model for each class!
def OneVsAll(X,y,num_labels,lmda):
    m,n = X.shape
    all_theta = np.zeros((num_labels, n + 1))
    
    X = np.insert(X,0,1,axis=1)
    
    initial_theta = np.zeros((n+1,1))
    for c in range(0,10):
        label = 10 if c==0 else c
        result = spo.minimize(fun=lr_cost_function,x0=initial_theta, args=(X,(y==label).astype(int),lmda),method='TNC', jac=True)
        all_theta[c,:] = result.x
    return all_theta
```

```python
all_theta = OneVsAll(X,y,num_labels, 0.1)
```

```python
def predictOneVsAll(theta,X):
    m = X.shape[0]
    num_labels = theta.shape[0]
    
    p = np.zeros((m,1))
    
    X = np.insert(X,0,1,axis=1)
    
    p = np.argmax(h(X.dot(theta.T)),axis=1) # check for which classification the highest probability/likelyhood is achieved --> best prediction
    p[p==0] = 10
    return p
```

```python
pred = predictOneVsAll(all_theta, X)
print("Accuracy: {0:8.2f}%".format(np.mean(pred==y)*100))
```

# Forwardpropagation Neural Network

```python
# Load data
mat = scipy.io.loadmat("ex3data1.mat")
X = mat["X"]
y = mat["y"].ravel() # IMPORTANT! For some reason the code will not work if .ravel() is not used. I've yet to figure out exactly why this happens.
m,n = X.shape
```

```python
weights = scipy.io.loadmat("ex3weights.mat") # import pretrained weights from matlab data file
```

```python
Theta1 = weights["Theta1"]
Theta2 = weights["Theta2"]
```

```python
# parameter setup
input_layer_size  = 400
hidden_layer_size = 25
num_labels = 10;
```

```python
def predictNN(Theta1, Theta2, X):
    m = X.shape[0]
    num_labels = Theta2.shape[0]
    p = np.zeros((m,1))
    
    a1 = np.insert(X,0,1,axis=1) # add x0 column to inputs
    z2 = a1.dot(Theta1.T) # calculate first input transformation
    a2 = np.insert(h(z2),0,1,axis=1) # insert a2_0 column to hidden_layer
    z3 = a2.dot(Theta2.T) # calculate hidden layer
    a3 = h(z3) # calculate final output layer
    p = np.argmax(a3,axis=1)
    p+=1 # adjust indices because matlab starts at index 1
    return p
```

```python
pred = predictNN(Theta1, Theta2, X)
```

```python
print("Accuracy: {0:8.2f}%".format(np.mean(pred==y)*100))
```
